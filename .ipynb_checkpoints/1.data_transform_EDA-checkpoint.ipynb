Do an exploratory data analysis of the adult dataset. The cell below loads the dataset from UCI Machine Learning Repository into a pandas dataframe called adult_data. It requires that you have installed the package ucimlrepo.

Fetching the data set and displaying the variables

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.mosaicplot import mosaic
```

```python
from ucimlrepo import fetch_ucirepo 
adult = fetch_ucirepo(id=2) 
X = adult.data.features 
y = adult.data.targets 
X["income"] = y
adult_data = X.copy()

df= adult_data


adult_data
```

```python
#Checking the amount of entries, columns, data types and null values of the adult data set. 
#Here we can see that 3 of the columns does indeed contain null values
adult_data.info()
```

```python
# Check for specific amount of missing values 
print("Missing values:\n", df.isnull().sum())
```

```python
# Drop rows with missing values
df.dropna(inplace=True)
```

```python
#Describing categorical values to show the amount of values, uniques and most frequent values
df.describe(include=[object])
```

```python
#Describing numerical values to get an overview of general distribution, the mean etc..
df.describe()
```

<b> Since we are visualizing distribution of all columns we are going to loop through them. We are using a countplot for categorical values and histplot for numerical values

```python
# Generate descriptive statistics and visualizations
for column in df.columns:
    print(f"\nDescriptive Statistics for {column}:")
    display(df[column].describe(include='all'))
    
    plt.figure(figsize=(8, 4))
    
    if df[column].dtype == 'object':  # Categorical variable
        sns.countplot(y=df[column], order=df[column].value_counts().index)
        plt.title(f"Distribution of {column}")
    else:  # Numerical variable
        sns.histplot(df[column], bins=30, kde=True)
        plt.title(f"Histogram of {column}")
    
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.show()

print("Data exploration completed.")
```

From the diagrams above we get a lot of information about the data set. 
The age column is highly dominated by 20-50 year olds.

Workclass is higly dominated by people working in the private sector, to the point where it could be difficult to really say anything meaningful about the other sectors. 

Occupation seems to include a ? value, which could either be unknown og unoccupied

Race is highly dominated by White, though there also is a relevant amount og Black and Asian. Again can we really say anything meaningful about the Eskimo and Other races. Arguably these columns could be dropped for future research.

There is about twice as many Males as Females in the dataset. When comparing the two Sexes there should be some form of scaling involved

Hours per has a mean of about 40 hours worked pr week, and most people seem to work around these 40 hours.

Distribution of native-country seems to be heavily dominated by US. Again it could be argued that other values should be dropped, since there are so few datapoints.

Income seems to contain values that are the same but for some reason as a '.' included. We will fix this before doing further exploration on the dataset



```python
#fixing the income column
df['income'] = df['income'].replace({'<=50K.': '<=50K', '>50K.': '>50K'})
```

```python
## exploring correlation/variation between 2 variables
```

```python
# 1. Categorical vs. Categorical. Here we are looking to see if there is a difference in the income between men and females.
mosaic(df, ["sex", "income"])
plt.title("Mosaic plot for categorical variables sex and income")
plt.xlabel("sex")
plt.ylabel("income")
plt.show()
```

<b> Explanation

The width of the blocks for "Male" and "Female" on the x-axis represents the proportion of males and females in the dataset.The Male section is larger than the Female section, meaning there are more males in the dataset.

The dataset shows a gender disparity in income, where a higher proportion of males earn more than 50K compared to females.
The majority of both sexes fall in the lower income category (<=50K), but the gender gap is visible in the higher-income segment.

```python
# 2. Numerical vs. Numerical. Here we are looking for a correlation between Hours Per Week and Age,
# to see if the the hours worked per week depends on the age of people in dataset
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['age'], y=df['hours-per-week'])
plt.title('Age vs. Hours Per Week')
plt.xlabel('Age')
plt.ylabel('Hours Per Week')
plt.show()
```

<b> Explanation

The majority of individuals seem to work between 30 to 50 hours per week, regardless of age.

Young adults (age ~18-30) exhibit high variability in work hours, with some working part-time (below 20 hours) and others working full-time or overtime.

Middle-aged individuals (age ~30-60) appear to have more consistent working hours, mostly clustering between 40-50 hours per week.

Older individuals (age ~60+) generally work fewer hours, with more variability in the number of hours worked.

The scatter plot suggests that age does not strongly determine how many hours a person works per week.
Most people work around 40 hours per week, though younger and older individuals have more variability in their work hours.

```python
# 3. Categorical vs. Numerical
plt.figure(figsize=(12,5))
sns.boxplot(x=df['education'], y=df['hours-per-week'])
plt.title('Education vs. Hours Per Week')
plt.xlabel('Education')
plt.ylabel('Hours Per Week')
plt.xticks(rotation=45)
plt.show()
```

<b> Explanation

Most education levels show a similar median workweek around 40 hours, suggesting that education does not drastically impact the typical number of hours worked. Doctorate and Professional School graduates have a higher median workweek compared to others.

There are many outliers above 60-80 hours per week, likely representing individuals working overtime or in high-demand jobs.
Some individuals report working close to 0 hours per week, possibly part-time workers or those unemployed.

Individuals with lower education levels (5th-6th grade, 7th-8th, etc.) show more variability in their work hours.
Higher education levels (e.g., Doctorate, Professional School) show more consistent work hours with fewer extreme outliers.
